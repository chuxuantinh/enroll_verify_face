```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
threshold = 0.4
thresholdNonBf = 260000
objects = ['wrong1', 'wrong2', 'wrong3', 'wrong4', 'correct', 'wrong5']
y_pos = ['wrong1', 'wrong2', 'wrong3', 'wrong4', 'correct', 'wrong5']
distance_BF = [0.45, 0.44, 0.43, 0.42, 0.31, 0.44]
distance_nonBF = [296799, 286218, 279329, 273329, 201500, 286842]
```

```python
plt.bar(y_pos, distance_BF, width=0.7, align='center', alpha=1, color='b')
plt.xticks(y_pos, objects)
plt.ylabel('Hamming Distance')
plt.title('Distance Evaluation - Threshold : 0.4')
plt.show()
```

```python
plt.bar(y_pos, distance_nonBF, width=0.7, align='center', alpha=1, color='g')
plt.xticks(y_pos, objects)
plt.ylabel('Hamming Distance')
plt.title('Distance Evaluation - Threshold : 26000')
plt.show()
```

```python
# data to plot
n_groups = 6
n_scale = thresholdNonBf / threshold
distance_nonBF_scale = [distance_nonBF[0]/n_scale, distance_nonBF[1]/n_scale, distance_nonBF[2]/n_scale, distance_nonBF[3]/n_scale, distance_nonBF[4]/n_scale, distance_nonBF[5]/n_scale]

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.3
opacity = 0.8

rects1 = plt.bar(index, distance_nonBF_scale, bar_width,
alpha=opacity,
color='g',
label='Non BF')

rects2 = plt.bar(index + bar_width, distance_BF, bar_width,
alpha=opacity,
color='b',
label='BF')

plt.ylabel('Hamming Distance')
plt.title('BF/NonBF Comparison')
plt.xticks(index + bar_width, ('wrong1', 'wrong2', 'wrong3', 'wrong4', 'correct', 'wrong5'))
plt.legend()

plt.tight_layout()
plt.show()
```
